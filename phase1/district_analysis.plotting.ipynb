```python
import geopandas as gpd
import pandas as pd
import plotly.express as px
```

```python
# ==========================================================
# PATHS
# ==========================================================

GEOJSON = "geoBoundaries-IND-ADM2.geojson"
CLUSTERS = "output/data/district_cluster_assignments.csv"

# ==========================================================
# LOAD
# ==========================================================

gdf = gpd.read_file(GEOJSON)
clusters = pd.read_csv(CLUSTERS)

gdf["shapeName"] = gdf["shapeName"].str.strip()
clusters["district"] = clusters["district"].str.strip()

# ==========================================================
# DISTRICT NAME FIXES
# ==========================================================

name_map = {
    "Allahabad": "Prayagraj",
    "Faizabad": "Ayodhya",
    "Sant Ravi Das Nagar": "Bhadohi",
    "Lakhimpur Kheri": "Kheri",
}

clusters["district"] = clusters["district"].replace(name_map)
```

```python
# ==========================================================
# MERGE
# ==========================================================

up = gdf.merge(
    clusters,
    left_on="shapeName",
    right_on="district",
    how="inner",
)

print(f"Matched districts: {len(up)}")

missing = sorted(set(clusters["district"]) - set(up["district"]))
print("\nMissing districts:")
print(missing)
```

```python
from rapidfuzz import process, fuzz

geo_names = sorted(gdf["shapeName"].unique())

missing = [
    "Amroha",
    "Ayodhya",
    "Barabanki",
    "Bhadohi",
    "Hathras",
    "Kasganj",
    "Maharajganj",
    "Prayagraj",
    "Shamli",
    "Shravasti",
    "Siddharth Nagar",
]

for name in missing:
    print(f"\n{name}")
    matches = process.extract(
        name,
        geo_names,
        scorer=fuzz.WRatio,
        limit=5,
    )
    for m in matches:
        print("   ", m)
```
